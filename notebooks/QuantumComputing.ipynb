A form of computing known as quantum computing is based on the ideas of quantum mechanics, a theory that explains how matter and energy behave at the atomic and subatomic scales.

Information is kept in bits in traditional computing, which can either be 0 or 1. Information is kept in quantum bits, or qubits, which can simultaneously represent 0 and 1. These bits are used in quantum computing. This characteristic, referred to as superposition, enables quantum computers to carry out specific sorts of calculations far more quickly than conventional computers.

Before they can be utilized extensively, quantum computers must first overcome a number of technological obstacles because they are still in the early phases of development. They have numerous potential uses in industries including medical, finance, and materials research, and they have the ability to tackle some problems far more quickly than conventional computers.

### H Gate

***

$ H = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix} \qquad | 0 \rangle = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$

$ H | 0 \rangle = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix} \begin{bmatrix} 1 \\ 0 \end{bmatrix} = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix}$

$P(0) = |\frac{1}{\sqrt{2}}|^2 = \frac{1}{2}$

$P(1) = |\frac{1}{\sqrt{2}}|^2 = \frac{1}{2}$

```python
# Create a circuit.
circuit = qiskit.QuantumCircuit(1, 1)

# Add a gate.
circuit.h(0)

# Measure.
circuit.measure(0, 0)

# Draw the circuit
circuit.draw(output='mpl')
```

```python
# Get a simulator.
sim = qiskit.Aer.get_backend('aer_simulator')

# Simulte the circuit.
result = sim.run(circuit).result().get_counts()

# Histogram.
viz.plot_histogram(result, figsize=(3, 3))
```

### Y Gate

***

$ Y = \begin{bmatrix} 0 & -i \\ i & 0 \end{bmatrix} \qquad | 0 \rangle = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$

$ Y | 0 \rangle = \begin{bmatrix} 0 & -i \\ i & 0 \end{bmatrix} \begin{bmatrix} 1 \\ 0 \end{bmatrix} = \begin{bmatrix} 0 \\ i \end{bmatrix}$

$P(0) = |0|^2 = 0$

$P(1) = |0+1i|^2 = 1$

```python
# Create a circuit.
circuit = qiskit.QuantumCircuit(1, 1)

# Add a gate.
circuit.y(0)

# Measure.
circuit.measure(0, 0)

# Draw the circuit
circuit.draw(output='mpl')
```

## Exercise


Adapt random_f to generate a random function that takes 3 bits (0 or 1 values) as input and returns a single bit.

***

```python
import random

def random_f():
    # generate a random integer in the range [0, 7]
    # this will represent the function's truth table
    truth_table = random.randint(0, 7)

    def f(x, y, z):
        # use bitwise AND to determine if the truth table value
        # at the index corresponding to the input values is 1
        return truth_table & (1 << (x*4 + y*2 + z)) > 0

    return f
```

## Exercise

Adapt random_f to generate a random function that is either constant or balanced, that takes 3 bits (0 or 1 values) as input and returns a single bit.

***

```python
import random

def random_f():
    # generate a random integer in the range [0, 1]
    # this will determine whether the function is constant (0) or balanced (1)
    func_type = random.randint(0, 1)

    if func_type == 0:
        # the function is constant, so generate a random output value
        output = random.randint(0, 1)

        def f(x, y, z):
            return output

        return f
    else:
        # the function is balanced, so generate a random truth table
        truth_table = random.randint(0, 7)

        def f(x, y, z):
            # use bitwise AND to determine if the truth table value
            # at the index corresponding to the input values is 1
            return truth_table & (1 << (x*4 + y*2 + z)) > 0

        return f
```
