# Fourier Transform Notebook
A mathematical technique called the Fourier transform is used to break down a function into its separate frequency components. It bears the name of Jean-Baptiste Joseph Fourier, a French mathematician who first proposed the idea in the early 19th century.

A signal or an image that is a function of time, such as the Fourier transform, is typically represented as a function of frequency. This can be helpful for filtering particular frequency components or for examining the frequency composition of a transmission.

Numerous industries, including engineering, physics, and computer science, use the Fourier transform. It is a key component of signal processing and has a wide range of uses, such as the processing of images and sounds, communication systems, and data analysis.

```python
# In Object Oriented form.
class Complex():
    def __init__(self, a, b):
        self.a = a
        self.b = b
    
    def __add__(self, other):
        return Complex(self.a + other.a, self.b + other.b)
    
    def __mul__(self, other):
        a, b = self.a, self.b
        c, d = other.a, other.b
        return Complex(a*c - b*d, a*d + b*c)
    
    def __repr__(self):
        return f'Complex({self.a}, {self.b})'
```

## Inverse Fourier Transform
The mathematical operation that is the inverse of the Fourier transform is called the inverse Fourier transform. It is employed to create a function from a set of frequency components.

The definition of the inverse Fourier transform, represented by the symbol F-1, is as follows:

f(t) = (1/2)  F()e^(it)d

The original function is denoted by f(t), its Fourier transform is F(), and the frequency is.

A function can be recovered from its Fourier transform using the inverse Fourier transform, or a function can be created by combining sinusoidal components at various frequencies. It is an essential instrument for signal processing and has several uses in physics, engineering, and other disciplines.

```python
# Invert the previous Fourier transform applied to x.
yinv = fft.ifft(y)
yinv
```

```python
Expected output:

array([ 1.0+0.j,  2.0+0.j,  1.0+0.j, -1.0+0.j,  1.5+0.j])
We expect x and yinv to be equal.
```

## Formula
$$
y[k] = \sum_{n=0}^{N-1} e^{-2 \pi j \frac{kn}{N}} x[n]
$$

```python
# Calculate by hand.

# Output array.
y = []

N = len(x)

for k in range(N):
    y_k = 0.0
    for n in range(N):
        y_k = y_k + (np.e**(-2 * np.pi * 1j * k * n / N) * x[n])
    y.append(y_k)
    
np.array(y)
```

```python
import scipy

import matplotlib.pyplot as plt
```

```python
# Number of samples.
N = 600

# Spacing between samples.
T = 1.0 / 800.0

# Range of x (input) values.
x = np.linspace(0.0, N*T, N, endpoint=False)

# Have a look at x.
x
```

```python
# First sine wave.
y_0 = np.sin(50.0 * 2.0 * np.pi * x)

# Plot it.
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y_0);
```

```python
# First sine wave.
y_1 =  0.5 * np.sin(80.0 * 2.0 * np.pi * x)

# Plot it.
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y_1);
```

```python
# Sum of the waves.
y = y_0 + y_1

# Plot it.
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y);
```

```python
# Fast Fourier transform of y.
yf = scipy.fft.fft(y)

# Frequencies for y.
xf = scipy.fft.fftfreq(N, T)[:N//2]
```
