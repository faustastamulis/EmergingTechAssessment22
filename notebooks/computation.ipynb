## Growth Rates
Growth rates in computing can refer to the rate at which an algorithm's time or spatial complexity develops as the size of the input increases. While space complexity refers to the amount of memory or storage space an algorithm needs, time complexity refers to the length of time it takes for an algorithm to run.
Consider the following two methods for sorting a list of numbers:

The time required to run Algorithm A grows at a rate of n2 as the input size (n) increases since it has a time complexity of O(n2).
The time it takes to perform Algorithm B grows at a rate of nlog(n) as the input size rises since the algorithm has a time complexity of O(nlog(n)).

In this situation, algorithm B would be more effective for higher input sizes and has a faster growth rate. It is crucial to keep in mind that time and space complexity frequently trade off, and that it is not always viable to reduce both at once.

```python
# Numbers.
import numpy as np

# Plotting.
import matplotlib.pyplot as plt

# Create a figure.
fig, ax = plt.subplots(figsize=(12, 6))

# x values.
x = np.linspace(0.0, 10.0, 1000)

# Plot polynomial.
ax.plot(x, x**2, label="$x^2$")

# Plot exponential.
ax.plot(x, 2**x, label="$2^x$")

# Legend.
ax.legend()

# Show.
plt.show();
```

## Big O Notation
The upper bound of the time or space complexity of an algorithm can be expressed using the ig O notation. It gives a ballpark approximation of how an algorithm's execution time or memory requirements increase with input size.

Big O notation is used to indicate the number of operations an algorithm executes as a function of the input size in the context of time complexity. For instance, if an algorithm has a time complexity of O(n), it means that as the size of the input rises, the algorithm performs n times more operations. If an algorithm has an O(n2) time complexity,this means that the number of operations grows at a rate of n^2 as the input size increases.

Describe and plot five examples of pairs of functions f and g such that f is O(g).

Here are five examples of pairs of functions f and g such that f is O(g):

* f(x) = 2x, g(x) = x
This is a linear relationship, where f grows at the same rate as g.

* f(x) = 10x^2 + 3x + 4, g(x) = x^2
This is a quadratic relationship, where f grows at a faster rate than g.

* f(x) = log(x), g(x) = x
This is a logarithmic relationship, where f grows at a slower rate than g.

* f(x) = sqrt(x), g(x) = x
This is a square root relationship, where f grows at a slower rate than g.

* f(x) = x^x, g(x) = x^2
This is an exponential relationship, where f grows at a faster rate than g.

Note that in all of these examples, the growth rate of f is upper-bounded by the growth rate of g. This means that there exists a constant c such that f(x) <= c*g(x) for all x greater than some threshold. This is the definition of f being O(g).

## Turing Machine
A Turing machine is an abstract machine that manipulates symbols on a strip of tape in accordance with a set of rules. It is a mathematical model of computation. A Turing machine, despite its simplicity, may be modified to imitate the logic of any computer method. It is particularly helpful in illustrating how a CPU works inside of a computer. Alan Turing, regarded as the founder of computer science, created the Turing machine in 1936.

```python
# State table.
table = {
    ('X', '_'): ['_', 'R', 'T'],
    ('X', '0'): ['0', 'R', 'X'],
    ('X', '1'): ['1', 'R', 'Y'],
    ('Y', '_'): ['_', 'R', 'F'],
    ('Y', '0'): ['0', 'R', 'Y'],
    ('Y', '1'): ['1', 'R', 'X'],
}

# Tape input.
tape = list('0101111')
# Position on tape.
pos = 0
# Initial state is first in table.
state = 'X'

# Keep going while we are not in a halting state.
while state not in ['T', 'F']:
    # Print the current status.
    print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
    # Get the row of the table.
    row = table[(state, tape[pos])]
    # Overwrite the symbol.
    tape[pos] = row[0]
    # Move left or right.
    if row[1] == 'R':
        # Put blanks on tape as necessary.
        if pos == len(tape) - 1:
            tape = tape + ['_']
        # Increase position.
        pos = pos + 1
    else:
        # Put blanks on tape as necessary.
        if pos == 0:
            tape = ['_'] + tape
            # The position on the tape has to move with it.
            pos = pos + 1
        # Decrease position.
        pos = pos - 1
    # Update the state.
    state = row[2]

# Print the current status.
print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
```

## Double Slit Experiment
A well-known physics experiment that illustrates the wave-particle duality of matter and the fundamentals of quantum mechanics is the double-slit experiment. It entails firing a stream of particles (such photons or electrons) through two slits that are close to one another and examining the interference pattern that is created on a screen behind the slits.

In the experiment, it is shown that the particles act like waves, creating an interference pattern on the screen as if they had simultaneously passed through both slits. The expected behavior of classical particles would be for them to pass through either one slit or the other and leave behind two separate bands on the screen, which is in contrast to this behavior.

Many theoretical and experimental studies have been conducted on the double-slit experiment, which has been demonstrated with a variety of particles. It has been the focus of considerable debate and discussion in the disciplines of physics and philosophy and is frequently used as a crucial example of how bizarre and counterintuitive quantum mechanics can be.

## Stern–Gerlach
Otto Stern and Walther Gerlach carried out the well-known Stern-Gerlach experiment in physics in 1922. It was created to show how silver atoms' magnetic moments may be quantized, and it has significantly advanced our knowledge of quantum mechanics.

A beam of silver atoms is subjected to the experiment's non-uniform magnetic field. The atoms' magnetic moments align with the magnetic field as a result of the magnetic field, and as a result, it is anticipated that the atoms will travel across the field in a curved fashion. However, the experiment showed that instead of the anticipated continuous curve, the atoms actually follow discontinuous routes that correspond to various magnetic moment values. The only way to explain this behavior is to assume that the magnetic moment of the atoms is a discrete variable, not a continuous variable as expected by classical physics.

The Stern-Gerlach experiment has had a tremendous impact on our knowledge of the nature of matter and energy and is frequently mentioned as a key illustration of the concepts of quantum mechanics.
